```python
import os
import shutil
import random

# Elbise veri seti yolu
dataset_path = "C://Users//emine//OneDrive//Masaüstü//catanddog//elbise"

# Train ve test veri seti yolları
train_path = "path/to/train_data/elbise"
test_path = "path/to/test_data/elbise"

# Train ve test klasörlerini oluştur
os.makedirs(train_path, exist_ok=True)
os.makedirs(test_path, exist_ok=True)

# Tüm sınıfları al
classes = os.listdir(dataset_path)

# Her sınıf için train ve test klasörlerini oluştur ve görüntüleri kopyala
for class_name in classes:
    class_dir = os.path.join(dataset_path, class_name)
    train_class_dir = os.path.join(train_path, class_name)
    test_class_dir = os.path.join(test_path, class_name)
    os.makedirs(train_class_dir, exist_ok=True)
    os.makedirs(test_class_dir, exist_ok=True)
    
    # Sınıfa ait tüm görüntülerin listesini al
    images = os.listdir(class_dir)
    
    # Görüntülerin %80'ini train, %20'sini test olarak ayır
    num_train_images = int(len(images) * 0.8)
    train_images = random.sample(images, num_train_images)
    test_images = [image for image in images if image not in train_images]
    
    # Train görüntülerini kopyala
    for image in train_images:
        src = os.path.join(class_dir, image)
        dst = os.path.join(train_class_dir, image)
        shutil.copy(src, dst)
        
    # Test görüntülerini kopyala
    for image in test_images:
        src = os.path.join(class_dir, image)
        dst = os.path.join(test_class_dir, image)
        shutil.copy(src, dst)
```
